```python
import json  

# Define the path to your JSON file  
json_file_path = 'xsections_had1.json'  # Replace with your actual file path
target = 'xsections_had.json'

# Open and load the JSON file  
with open(json_file_path, 'r') as file:  
    data = json.load(file)  

# Modify the keys in the dictionary  
modified_data = {}  
for key in data:  
    # Replace 'MINIAODSIM' with 'NANOAODSIM' in the key  
    new_key = key.replace('MiniAODAPVv2', 'NanoAODAPVv9')  
    # Update the dictionary with the new key  
    modified_data[new_key] = data[key]  

# Overwrite the original JSON file with the modified data  
with open(target, 'w') as file:  
    json.dump(modified_data, file, indent=4)  

print("JSON file has been updated successfully.")  
```
